# Introductory Commands

These were the commands used in lecture to help introduce the colleges and narrow down the dataframes.

```python
import pandas as pd
college = pd.read_csv("../../college.csv", index_col="INSTNM")
# college
nc = college[college["STABBR"]=='NC']
# nc
```

# 1a) Conditions:
- Undergraduates > 7000

- Sorted descendingly by % receiving federal loans

## Question: How many of the top 50 are located in NC?

Step-wise Process: I conversed with Dr. Whaley to grab the conditions from all the colleges then pull the North Carolinian colleges from that subgroup.

```python
plus7k = college[college["UGDS"]>=7000] # First condition
# plus7k.columns
sorted7k = plus7k.sort_values("PCTFLOAN",ascending=False).head(50) # Second condition 
ncplus7k = sorted7k[sorted7k["STABBR"] == 'NC']
ncplus7k # Only one school from the top 50 is located in NC
```

### Only 1 North Carolinian college is in the top 50 category.

# 1b) Conditions:
- Undergraduates > 7000

- Sorted descendingly by % receiving federal loans

## Question: How many of the top 50 are distance only?

Step-wise Process: Similar to last problems except for the last distance condition.

```python
plus7k = college[college["UGDS"]>=7000] # First condition
sorted7k = plus7k.sort_values("PCTFLOAN",ascending=False).head(50) # Second condition 
dist = sorted7k[sorted7k["DISTANCEONLY"]==1.0] # Distance only is True/1.0
dist
```

### 5 Colleges in the top 50 are distance only.

# 1c) Conditions:
- Undergraduates > 7000

- Sorted descendingly by % receiving federal loans

## Question: What is the mean earning of students 10years after enrollment of those schools?

Step-wise Process: Similar to last problems except for removing the exceptions and calculating the mean earnings.

```python
plus7k = college[college["UGDS"]>=7000] # First condition
sorted7k = plus7k.sort_values("PCTFLOAN",ascending=False).head(50) # Second condition 
# sorted7k["MD_EARN_WNE_P10"]!= "PrivacySuppressed"
dropped = sorted7k[sorted7k["MD_EARN_WNE_P10"]!= "PrivacySuppressed"] # Drop the private numbers
dropped = dropped.dropna(subset=["MD_EARN_WNE_P10"]) # Drop the NaN values
earnings = dropped["MD_EARN_WNE_P10"].astype(float).mean() # Finally take the mean
earnings
```

### $40027.083333333336 is the mean earning from the top 50 schools.

# 2) What is the national median debt for graduates (GRAD_DEBT_MDN_SUPP) compared to the median for NC schools?

Step-wise Process: Get rid of the private and NaN values for all colleges, then for NC.

```python
dropped = college[college["GRAD_DEBT_MDN_SUPP"]!= "PrivacySuppressed"] # Drop the private numbers
dropped = dropped.dropna(subset=["GRAD_DEBT_MDN_SUPP"]) # Drop the NaN values
natMeanDebt = dropped["GRAD_DEBT_MDN_SUPP"].astype(float).median()
nc = college[college["STABBR"] == "NC"]
droppednc = nc[nc["GRAD_DEBT_MDN_SUPP"]!= "PrivacySuppressed"] # Drop the private numbers
droppednc = droppednc.dropna(subset=["GRAD_DEBT_MDN_SUPP"]) # Drop the NaN values
ncMeanDebt = droppednc["GRAD_DEBT_MDN_SUPP"].astype(float).median()
print("The national median debt is \n"+ str(natMeanDebt)+ "\nThe NC median debt is \n"+ str(ncMeanDebt))
```

The North Carolinian median debt is higher than the national median debt, somehow.
