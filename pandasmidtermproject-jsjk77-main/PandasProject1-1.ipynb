# First have to import libraries and the file:

```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
confirmed = pd.read_csv("./time_series_covid19_confirmed_US.csv", index_col="UID")
confirmed
```

## Also: Subcategorical dataframe to states/provinces:

```python
# confirmed.columns
byState = confirmed.groupby("Province_State").sum()
byState
```

## Section 1, Question A:
Which state/territory had the highest numerical increase in confirmed cases over the summer months (July 1st through the end of September)

Note: Utilized NumPy and Pandas

```python
july1 = np.argmax(byState.columns == "7/1/20")

# up to, but not including/through the end of September
oct1 = np.argmax(byState.columns == "10/1/20") 
frame = byState.iloc[:,july1:oct1]
frame
```

### Answer:

```python
end = frame.iloc[:,-1]
start =  frame.iloc[:,1]
diff = end-start
loc = np.argmax(diff) # locates which row/state had the highest
finalstate = frame.iloc[loc]
finalstate
```

#### Texas had the highest numerical increase in confirmed cases over the summer months (July 1st through the end of September).

## Section 2, Question A Plot: 

Note: Used plotting functions with Pandas dataframe

```python
new = byState.iloc[:,july1:oct1] # Copying to not edit state subframe
new.loc[:,"Diff"] = diff
new = new.iloc[:,-1:]
new.sort_values(by="Diff",ascending=False,inplace=True)
top10ci = new.head(10).plot.bar()
top10ci.figure.suptitle("Top 10 States with Highest Case Increases")
top10ci.set_ylabel("# of COVID cases")
```

## Section 1, Question B:
Which state/territory had the highest percentage increase in confirmed cases over the same time period? 

Note: Same process as Question A, but with %

### Answer:

```python
percentage = (end-start)/start*100 
loc2 = np.argmax(percentage) # Locating the state/territory
frame.iloc[loc2]
```

### The Virgin Islands had the highest percentage increase in confirmed cases over the same time period.

## Section 2, Question B Plot: 

Note: Same as Question A above

```python
new2 = byState.iloc[:,july1:oct1] # Copying to not edit state subframe
new2.loc[:,"Percentage"] = percentage
new2 = new2.iloc[:,-1:]
new2.sort_values(by="Percentage",ascending=False,inplace=True)
top10p = new2.head(10).plot.bar()
top10p.figure.suptitle("Top 10 States with Highest Percentage Increases")
top10p.set_ylabel("% Increase in COVID cases")
```

## Section 1, Question C:
What was the 7-day average increase in cases per state? (Mean of cases per day over a seven day period for 8 weeks, starting on June 1st)

Note: Attempted to count and average based on "per day" for 8 total weeks from June 1 - June 7, June 8 - 14, ... , July 20 - 26

### Answer:

```python
june1 = np.argmax(byState.columns == "6/1/20")
july27 = np.argmax(byState.columns == "7/27/20") # up to not including 8 weeks from June 1 (July 26)
frame3 = byState.iloc[:,june1:july27]
# len(frame3.columns) # 56 = 8 weeks * 7 days
temp = pd.DataFrame()
count = 1 # week # tracker
for i in frame3:
    if count%7 == 1: # Beginning of the week
        start = frame3[i]
    if count%7==0: # End of the week
        end = frame3[i]
        frame3.loc[:,count] = (end-start)/7 # Difference per day
    count+=1
mean = frame3.iloc[:,-8:] # Mean across the states
mean
```

## Section 2, Question C Plot: 

Note: Same as Question A and B

```python
q3 = byState.iloc[:,june1:july27]
for i in range(len(mean)//7):
    q3.loc[:,"Week " + str(i+1)] = mean.iloc[:,i]
# Tagging on the means to the end of the dataframe
q3 = q3.iloc[:,-8:]
q3
```

```python
top10q3 = q3.head(6).plot.bar()
top10q3.figure.suptitle("First 6 States' Weekly Increases")
top10q3.set_ylabel("# in COVID cases")
```

## Section 1, Question D:
What proportion of cases nationally had occurred in NY State at the start of each month, from March 2020 to August 2020? 

Note: Found cases for the start of each month for New York, then for the US as a whole (nationally)

```python
march1 = np.argmax(byState.columns == "3/1/20")
august1 = np.argmax(byState.columns == "9/1/20") # up to not including September 1 (End of August)
frame4 = byState.iloc[:,march1:august1]
ny = frame4.loc["New York"]
nyfiltered = ny.filter(regex="[345678]/1/20") # Specifically picks the start of the months
nyfiltered
```

```python
# For the rest of the country
us = confirmed.groupby("Country_Region").sum()
frame4b = us.iloc[:,march1:august1]
usa = frame4b.loc["US"]
usafiltered = usa.filter(regex="[345678]/1/20")
usafiltered
```

### Answer:

```python
proportion = nyfiltered/usafiltered
proportion
```

## Section 2, Question D Plot: 

Note: Couldn't do regular slicing on proportion for some reason (proportion[:,0] didn't work), so I did it this way (similar to lecture)

```python
# Organizing Data
nyp = []
otherp = []
for i in proportion:
    nyp.append(i)
    otherp.append(1-i)
nyp
```

```python
# Created new dataframe to plot
df = pd.DataFrame(index = ["3/1/20","4/1/20","5/1/20","6/1/20","7/1/20","8/1/20"])
df["New York"] = nyp
df["Other States"] = otherp
# df.rename(index={0: "3/1/20",1:"4/1/20",2:"5/1/20",3:"5/1/20",4:"6/1/20",5:"7/1/20"})
df
```

```python
plot = df.plot(kind="area")
plot.figure.suptitle("Proportions of Confirmed Cases of New York Nationally")
plot.set_xlabel("Month")
plot.set_ylabel("Proportion")
```
