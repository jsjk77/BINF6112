# Function, For Loop, and Finalization

My thought process:

1. Load in the FASTA (.fa) file.

2. Separate and search for all of the contents.

3. Put into temporary python dictionary.

4. Transfer data into pandas dataframe.

```python
import pandas as pd
import numpy as np

# My old UNC COMP 555 code to load .fa files
def loadFasta(filename):
    fp = open(filename, 'r')
    # split at headers
    data = fp.read().split('>')
    fp.close()
    # ignore whatever appears before the 1st header
    data.pop(0)     
    headers = []
    sequences = []
    for sequence in data:
        lines = sequence.split('\n')
        headers.append(lines.pop(0))
        sequences.append(''.join(lines))
    return (headers, sequences)
header, seq = loadFasta("./Mdomestica-pandasLab.fa")
# header[0].split()

# For loop to create all the corresponding values
label = []
seqLen = []
gc = []
sc = []
ec = []
frame = []
count = 0
for num in header:
    # Sequence Label
    label.append(num[:12])
    
for i in seq:
    # Sequence Length
    seqLen.append(len(i))
    
    # Start Codon
#     print(i[:3])
    if i[:3] == "ATG":
        sc.append(True)
    else:
        sc.append(False)
        
    # End Codon
#     print(i[-3:])
    if i[-3:] == 'TGA' or i[-3:] == 'TAG' or i[-3:] == 'TAA':
        ec.append(True)
    else:
        ec.append(False)
        
    # Complete Frame
    if len(i)%3 == 0:
        frame.append(True)
    else:
        frame.append(False)
        
    # GC Content
    count = 0
    for j in i:
#         print(j)
        if (j == 'G') or (j == 'C'):
            count += 1
    gc.append(count/len(i)*100)

# Created a temporary dictionary to transfer to pandas dataframe
tempDict = {'Sequence Label': label, 'Length': seqLen, 'GC Content (%)': gc, 'Start Codon': sc, 'End Codon': ec,'Complete Frame': frame}

# Conversion to pandas dataframe
df = pd.DataFrame(tempDict, columns=list(tempDict.keys()))
df
```
